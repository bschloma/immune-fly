# Batch run for a bunch of datasets

```python
experiment_paths = [
    r'/media/brandon/Data2/Brandon/fly_immune/Lightsheet_Z1/2023_05_24-dpt-gfp_r4-gal4_uas-mcd8-mcherry_no_inj_early_mid_24hrs',
    r'/media/brandon/Data2/Brandon/fly_immune/Lightsheet_Z1/2023_05_24-dpt-gfp_r4-gal4_uas-mcd8-mcherry_mock_inj_early_mid_24hrs']
zarr_paths = []
for path in experiment_paths:
    larvae_dirs = glob.glob(path + '/larvae*')
    for larvae_dir in larvae_dirs:
        zarr_paths.append(Path(larvae_dir) / 'im.ome.zarr/0')

model_name = 'confocal_2D_unet_bce_dice_ds3x'
for i, path_to_zarr in enumerate(zarr_paths):
    if not path_to_zarr.is_dir():
        continue
    try:
        pred_path = path_to_zarr.parent.parent / 'prediction.zarr'
        im = zarr.open(path_to_zarr, 'r')
        pred = zarr.open(pred_path, mode='w', shape=im[0,1].shape, chunks=im.chunks[2:], dtype='float32')

        for z in range(pred.shape[0]):
            this_slice = im[0, 1, z]
            this_pred = unet_predictions(this_slice, model_name=model_name, patch=(1, 800, 1920), device='cuda')
            pred[z] = this_pred.squeeze()
        print(f'finished {i+1} of {len(zarr_paths)} datasets')
    except Exception(f'error with {path_to_zarr}'):
        continue
```
