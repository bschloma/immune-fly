```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_cells = 10000
xc = 1
x = np.linspace(0.01,1,1000)
E = 1 / x #np.exp(-x / xc)
pos = np.random.uniform(size=num_cells)
E_at_pos = np.interp(pos, x, E)
T = 24
gamma = 1
r = 10
n = 1

A = (1 - np.exp(-gamma * T)) * r * E_at_pos ** n / gamma 
```

```python
bins = np.logspace(2, 3, 50+1)
#bins = np.logspace(np.log(np.min(A)), np.log(0.1 * np.max(A)), 100+1)
counts, bins = np.histogram(A, bins)
binwidths = np.diff(bins)
probdens = counts / np.sum(counts) / binwidths

plt.figure()
plt.plot(bins[:-1], probdens)
plt.xscale('log')
plt.yscale('log')
plt.xlim([0.8 * np.min(bins), np.max(bins)])
#plt.ylim([1e-2, 1e1])
#plt.xlim([1e-2, 1e2])

```
